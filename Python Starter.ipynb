```python
import requests
import json
import pandas as pd
from bs4 import BeautifulSoup
start = 12600
end = 12710

def get_clinic_name(clini_id):
    url = f"https://{clini_id}.portal.athenahealth.com/"
    response = requests.get(url).text
    soup = BeautifulSoup(response,'html.parser')
    clinic_name = soup.find_all('h1')[-1].text.strip()
    return clinic_name

master_list = []
for clinic_id in range(start,end+1):
    data_dict = {}
    data_dict['clinic_id'] = clinic_id
    data_dict['clinic_name'] = get_clinic_name(clinic_id)
    if data_dict['clinic_name'] != 'Payment Confirmation' and data_dict['clinic_name'] != "Sorry, we can't find that practice. Make sure you typed the right address." and len(data_dict['clinic_name']) != 0:
            master_list.append(data_dict)
    
    print(clinic_id)
df = pd.DataFrame(master_list)
# df.to_csv('clinic_name.csv',index=False)






```

```python
import requests
import datetime
import uuid
import json
import logging
from bs4 import BeautifulSoup
from pprint import pprint

logging.basicConfig(
    filename='Brands.log',
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    datefmt="%Y-%m-%d %H:%M:%S"
)

def get_url_text():
    uuid4 = uuid.uuid4()
    nameList = []
    try:
        url = "http://pythonscraping.com/pages/warandpeace.html"
        request = requests.get(url).text
        bsObj = BeautifulSoup(request,'html.parser')
#         html_tags = <span class="green"></span>
        html_tags = bsObj.find_all("span",{"class":"green"})
        for name in html_tags:
            nameList.append(name.get_text().replace("\n", ""))
        logging.info(f"request:{url}-response:{response}") 
#         doc.find(text="INSERT FOOTER HERE").replace_with(footer)
        response = nameList
    except Exception as e:
        logging.error(f'An error occured at {uuid4} {e})\n\n')
        response = f'{e} An  error occurred while scrapping correlation-id: {uuid4}'
    return response


def get_exp_text():
    uuid4 = uuid.uuid4()
    nameList = []
    response = ""
    try:
        url = "http://pythonscraping.com/pages/page3.html"
        request = requests.get(url).text
        bsObj = BeautifulSoup(request,'html.parser')
#         html_tags = <span class="green"></span>DescriptionImage
        html_tags = bsObj.find_all("tr",{"class":"gift"})
        for name in html_tags:
            dic  =  {}
            dic['Title'] = name.find_all('td')[0].get_text().replace("\n"," ")
            dic['Description'] = name.find_all('td')[1].get_text().replace("\n"," ")
            dic['Price'] = name.find_all('td')[2].get_text().replace("\n"," ")
            dic['Image'] = name.find_all('td')[3].find('img',{"src"})
            nameList.append(dic)
        logging.info(f"request:{url}-response:{nameList}") 
        response = nameList
    except Exception as e:
        logging.error(f'An error occured at {uuid4} {e})\n\n')
        response = f'{e} An  error occurred while scrapping correlation-id: {uuid4}'
    return response        
  
print(json.dumps(get_exp_text()))

    
```
