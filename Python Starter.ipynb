```python
import requests
import datetime
import uuid
import json
import logging
import pandas as pd
from bs4 import BeautifulSoup
from pprint import pprint
import re
import random

from IPython.display import display, HTML
display(HTML("<style>.container { width:95% !important; }</style>"))

logging.basicConfig(
    filename='Brands.log',
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    datefmt="%Y-%m-%d %H:%M:%S"
)

def get_url_text():
    uuid4 = uuid.uuid4()
    nameList = []
    req_id = str(datetime.datetime.now().strftime("%Y%m%d%H%M%S%MS"))
    try:
        url = "http://pythonscraping.com/pages/warandpeace.html"
        request = requests.get(url).text
        bsObj = BeautifulSoup(request,'html.parser')
#         html_tags = <span class="green"></span>
        html_tags = bsObj.find_all("span",{"class":"green"})
        for name in html_tags:
            nameList.append(name.get_text().replace("\n", ""))
#         doc.find(text="INSERT FOOTER HERE").replace_with(footer)
        response = nameList
        logging.info(f"{req_id}-request:{url}-response:{response}") 
    except Exception as e:
        logging.error(f'{req_id}-An error occured at {uuid4} {e})\n\n')
        response = f'{e}-{req_id} An  error occurred while scrapping correlation-id: {uuid4}'
    return response


def get_exp_text():
    uuid4 = uuid.uuid4()
    nameList = []
    response = ""
    req_id = str(datetime.datetime.now().strftime("%Y%m%d%H%M%S%MS"))
    try:
        url = "http://pythonscraping.com/pages/page3.html"
        request = requests.get(url).text
        bsObj = BeautifulSoup(request,'html.parser')
#         html_tags = <tr class="gift"></span>DescriptionImage
        html_tags = bsObj.find_all("tr",{"class":"gift"})
        for name in html_tags:
            dic  =  {}
            dic['Title'] = name.find_all('td')[0].get_text().replace("\n"," ")
            dic['Description'] = name.find_all('td')[1].get_text().replace("\n"," ")
            dic['Price'] = name.find_all('td')[2].get_text().replace("\n"," ")
            dic['Image'] = name.find_all('td')[3].find('img').attrs['src']
            nameList.append(dic)
        logging.info(f"requestid:{req_id}-request:{url}-response:{nameList}") 
#         response = nameList
        response = json.dumps(nameList)
    except Exception as e:
        logging.error(f'requestid:{uuid4}-{req_id} An error occured at {e})\n\n')
        response = f'{e} requestid:{req_id} An  error occurred while scrapping correlation-id: {uuid4}'
    return response   



def get_link_text(article):
    uuid4 = uuid.uuid4()
    response = ""
    links = []
    try:
        url = "http://en.wikipedia.org/wiki/"+article
        request = requests.get(url).text
        bsObj = BeautifulSoup(request,'html.parser')
        #html_tags = <div id="bodyContent"></span> <a href="wiki"></a>
        for link in bsObj.find("div",{"id":"bodyContent"}).find_all("a",href=re.compile("^(/wiki/)((?!:).)*$")):
            if 'href' in link.attrs:
                links.append(link.attrs['href'])
        response = links        
    except Exception as e:
        logging.error(f'An error occured at {uuid4} {e})\n\n')
        response = f'{e} An  error occurred while scrapping correlation-id: {uuid4}'
    return response


def get_link_wiki_text(article):
    uuid4 = uuid.uuid4()
    response = ""
    try:
        url = "http://en.wikipedia.org/wiki/"+article
        request = requests.get(url).text
        bsObj = BeautifulSoup(request,'html.parser')
        response = bsObj.find("div",{"id":"bodyContent"}).find_all("a",href=re.compile("^(/wiki/)((?!:).)*$"))
    except Exception as e:
        logging.error(f'An error occured at {uuid4} {e})\n\n')
        response = f'{e} An  error occurred while scrapping correlation-id: {uuid4}'
    return response


pages = set()
def get_page_text(pageUrl):
    global pages
    uuid4 = uuid.uuid4()
    response = ""
    try:
        url = "http://en.wikipedia.org"+pageUrl
        request = requests.get(url).text
        bsObj = BeautifulSoup(request,'html.parser')
        #html_tags = <div id="bodyContent"></span> <a href="wiki"></a>
        for link in bsObj.find_all("a",href=re.compile("^(/wiki/)")):
            if 'href' in link.attrs:
                if link.attrs['href'] not in pages:
                    newPage = link.attrs['href']
                    print(newPage)
                    pages.add(newPage)
                    get_page_text(newPage)
        response = links        
    except Exception as e:
        logging.error(f'An error occured at {uuid4} {e})\n\n')
        response = f'{e} An  error occurred while scrapping correlation-id: {uuid4}'
    return response



# try:
#     uuid4 = uuid.uuid4()
#     links = get_link_wiki_text("kevin_Bacon")
#     now = datetime.datetime.now()
#     date_string = now.strftime("%Y%m%d%H%M%S")
#     date_int = int(date_string)
#     random.seed(int(datetime.datetime.now().strftime("%Y%m%d%H%M%S")))
#     while len(links) > 0:
#         newArticke = links[random.randint(0,len(links)-1)].attrs["href"]
#         print(newArticke)
#         links = get_link_wiki_text(newArticke)
# except Exception as e:
#         logging.error(f'An error occured at {uuid4} {e})\n\n')
#         print(f'{e} An  error occurred while scrapping correlation-id: {uuid4}') 
        
#########################################################################################################################################
# print(json.dumps(get_city_weather)) 
# get_link_text("kevin_Bacon")
# json.dumps(get_exp_text())
# df = pd.DataFrame(get_exp_text())
# df.to_csv('clinic_name.csv',index=False)
get_page_text("")
```
